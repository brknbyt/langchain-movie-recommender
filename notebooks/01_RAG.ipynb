```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("alanvourch/tmdb-movies-daily-updates")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(f"{path}/TMDB_all_movies.csv", )
df.info()
#data.describe().T
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
"title"
loader = CSVLoader(file_path=f"{path}/TMDB_all_movies.csv", source_column="title", content_columns=["title", "overview", "release_date", "runtime", "budget", "original_language", "original_title", "genres", "spoken_languages", "cast", "director", "director_of_photography", "writers", "music_composer"])

movie_documents = loader.load()

print(movie_documents[:5])  
print("Chunk size:", len(movie_documents[0].page_content))
```

```python
movie_documents[:5]
```

```python
from langchain_text_splitters import CharacterTextSplitter

char_text_splitter = CharacterTextSplitter(
    separator=".",
)

chunks = char_text_splitter.split_documents(movie_documents)

print(chunks[0])  # Print the first chunk to verify splitting
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vector_store = InMemoryVectorStore(embeddings)

document_ids = vector_store.add_documents(documents=chunks[:250])

print(document_ids[:5])
```

```python
from langchain_core.documents import Document

query = "Which movies included Danny Verduzco in their cast?"
retrieved_docs = vector_store.similarity_search_with_score(query, k=3, filter=None)
#vector_store.
serialized = "\n\n".join(
    (f"Source: {doc.metadata}\nContent: {doc.page_content}\nScore: {score}")
    for doc, score in retrieved_docs
)
#for doc in retrieved_docs:
#    print(doc, "\n")

print(serialized)
```
